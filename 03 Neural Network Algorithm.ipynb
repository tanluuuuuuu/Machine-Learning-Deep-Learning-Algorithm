```python
import numpy as np
from math import exp
```

```python
def MeanSquareError(y_true, y_predict):
    # y_true, y_predict: Two 1D np.array have the same size
    return np.mean((y_true - y_predict)**2)
```

```python
y = np.array([1, 2])
y_predict = np.array([1.5, 2.5])
print(MeanSquareError(y, y_predict))
```

```python
def DerivativeMeanSquareError(y_true, y_predict):
    return 2 * (y_predict - y_true) / len(y_true)
```

```python
y = np.array([1, 2])
y_predict = np.array([1.5, 2.5])
print(DerivativeMeanSquareError(y, y_predict))
```

```python
print(DerivativeMeanSquareError(y, y_predict))
```

```python
class Dense:
    def __init__(self, n_input, n_output, activation = None):
        self.weights = np.ones((n_output, n_input))
        self.bias = np.ones(n_output)
        self.input = None
        if (activation != None):
            activation = activation.lower().strip()
        self.activation = activation

    def forward(self, input):
        self.input = input
        output = np.dot(self.weights, input) + self.bias
        if (self.activation == None):
            return output
        elif (self.activation == 'sigmoid'):
            return np.array([1.0 / (1.0 + exp(-x)) for x in output])
        elif (self.activation == 'tanh'):
            return np.array([(exp(x) - exp(-x)) / (exp(x) + exp(-x)) for x in output])

    def backward(self, gradient, learning_rate = 0.000001):
        self.weights = self.weights - learning_rate * np.dot(gradient, self.input.T)
        self.bias = self.bias - learning_rate * gradient
        return np.dot(self.weights.T, gradient)
```

```python
x = Dense(2, 2, 'sigmoid')
fx = x.forward([7, 7])
print(fx)
```

```python
x = Dense(2, 2)
fx = x.forward([7, 7])
print(fx)
```

```python
x = Dense(2, 2, 'tanh')
fx = x.forward([7, 7])
print(fx)
```
