```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2 * np.random.rand(100, 1)
Y = 4 + 3*X + np.random.randn(100, 1)
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X, Y)
```

```python
X = np.c_[np.ones((100, 1)), X]
```

```python
theta_best = np.random.rand(2, 1)
for i in range(1000):
    theta_best = theta_best - 0.1 * 2/len(X) * X.T.dot(X.dot(theta_best) - Y) 
```

```python
print(theta_best)
```
