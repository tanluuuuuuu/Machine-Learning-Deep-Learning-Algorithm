## Import libraries

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
````

## Import dataset

````python
data = load_iris()
X = data.data
y = data.target
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=7)
````

## Build Linear Regression class

````python
class LinearRegressionLL:
    def __init__(self):
        self.w = None
        self.b = None
    
    def fit(self, X, y, learning_rate = 0.1, iterations = 10000):
        self.w = np.random.rand(X.shape[1], 1)
        self.b = np.random.rand(1, 1)
        n = X.shape[0]
        for _ in range(iterations):
            predictions = np.dot(X, self.w) + self.b
            dw = self.w - learning_rate * (2 / n) * X.T.dot(predictions - y)
            dw = self.b - learning_rate * (2 / n) * (predictions - y)

    def predict(self, X):
        return np.dot(X, self.w)
````

````python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

modelLL = LinearRegressionLL()
modelLL.fit(X, y)
````

````python
predictionsLL = modelLL.predict(X_test)
predictions = model.predict(X_test)
````

## Evaluate

````python
from sklearn.metrics import mean_squared_error
print('Sklearn model: ', mean_squared_error(y_test, predictions, squared=False))
print('Self model: ', mean_squared_error(y_test, predictionsLL, squared=False))
````
